## Observed Weather Forcing

Plot Medina Basin average temperature and precipitation for 2015-06-01 through 2020-01-01

```python
InFiler = os.path.normpath( os.path.join( IN_DIR1, "FBas_Medina_MetTS_1980through2020_DF.pkl" ) )
MedDF = pd.read_pickle( InFiler )
```

```python
display( HTML( MedDF.head().to_html() ) )
```

```python
# make monthly so that easier to see.
MedTempDF = MedDF[["AveT_C"]].copy()
MedPreDF = MedDF[["Pre_mm"]].copy()
```

```python
MonTempDF = MedTempDF.resample( 'MS' ).mean()
MonPreDF = MedPreDF.resample( 'MS' ).sum()
```

```python
MonTempDF = MonTempDF.loc[pd.Timestamp(2015, 6, 1, 0):pd.Timestamp(2020, 1, 1, 0 )].copy()
MonPreDF = MonPreDF.loc[pd.Timestamp(2015, 6, 1, 0):pd.Timestamp(2020, 1, 1, 0 )].copy()
```

```python
display( HTML( MonTempDF.describe().to_html() ) )
```

```python
display( HTML( MonPreDF.describe().to_html() ) )
```

```python
# add new indices
NewIndexer = [ x + pd.Timedelta(days=14) for x in MonPreDF.index.to_list() ]
```

```python
MonTempDF["NewIndexer"] = NewIndexer
MonPreDF["NewIndexer"] = NewIndexer
```

```python
MonTempDF.set_index( "NewIndexer", drop=True, inplace=True )
MonPreDF.set_index( "NewIndexer", drop=True, inplace=True )
```

```python
display( HTML( MonTempDF.head().to_html() ) )
```

```python
bWidth = 20.0
```

```python
xTicks = [ pd.Timestamp( 2016, 1, 1, ), pd.Timestamp(2017, 1, 1, ), pd.Timestamp(2018, 1, 1, ), pd.Timestamp(2019, 1, 1, ), pd.Timestamp(2020, 1, 1, ), ]
monTLabels = [ x.strftime("%Y-%m-%d") for x in xTicks ]
preTicks = [ 0.0, 100.0, 200.0, 300.0, 400.0, 500.0, 600.0 ]
tmpTicks = [ 5.0, 10.0, 15.0, 20.0, 25.0, 30.0, 35.0, ]
```

```python
OutSVG = os.path.normpath( os.path.join( OUT_DIR1, "ObsWeatherForcing.svg" ) )
OutPNG = os.path.normpath( os.path.join( OUT_DIR1, "ObsWeatherForcing.png" ) )
OutPDF = os.path.normpath( os.path.join( OUT_DIR1, "ObsWeatherForcing.pdf" ) )
```

```python
Fig1 = plt.figure()
Fig1.set_size_inches(7.0, 4.0)
ax11 = Fig1.add_subplot(111)
preRects = ax11.bar( MonPreDF.index, MonPreDF["Pre_mm"], bWidth, 
                     color='xkcd:ocean blue', edgecolor='xkcd:ocean blue',
                     label="Precipitation" )
ax11.set_xticks( xTicks )
ax11.set_yticks( preTicks )
ax11.set_xticklabels( monTLabels, minor=False, rotation=0 )
ax11.set_ylabel('Monthly Precipitation Depth ($mm$)', fontsize=10, color='xkcd:ocean blue')
ax11.set_xlim( (pd.Timestamp(2015, 7, 1), xTicks[len(xTicks)-1] ) )
ax11.set_ylim( (preTicks[0], preTicks[len(preTicks)-1]) )
ax11.tick_params(axis='both', which='major', labelsize=10)
ax11.tick_params(axis='y', colors='xkcd:ocean blue')
ax11.spines["right"].set_color('xkcd:ocean blue')
ax11.yaxis.set_major_formatter( mpl.ticker.StrMethodFormatter( "{x:,.0f}" ) )
# second plot
ax21 = ax11.twinx()
aveT, = ax21.plot( MonTempDF.index, MonTempDF["AveT_C"], linestyle='-', linewidth=2, color='xkcd:dark orange',
                   label="Mean Temperature" )
ax21.set_yticks( tmpTicks )
ax21.set_ylabel('Monthly Temperature ($ ^{\circ} C$)', fontsize=10, color='xkcd:burnt orange')
ax21.spines["right"].set_color('xkcd:burnt orange')
ax21.set_ylim( (tmpTicks[0], tmpTicks[len(tmpTicks)-1]) )
ax21.tick_params(axis='y', which='major', labelsize=10, colors='xkcd:burnt orange')
ax11.legend( (preRects, aveT, ), 
             ("Precipitation", "Mean Temperature" ),
             loc='upper center', fontsize=9, ncol=2 )
ax11.grid(visible=True, which='major', axis='y' )
ax21.yaxis.set_major_formatter( mpl.ticker.StrMethodFormatter( "{x:,.0f}" ) )
Fig1.savefig( OutSVG, dpi=600.0 )
Fig1.savefig( OutPNG, dpi=600.0 )
Fig1.savefig( OutPDF, dpi=600.0 )
```

## Annual Precip CN PDFs and Histograms

```python
InFiler = os.path.normpath( os.path.join( IN_DIR2, "CMIP5and6_ConeUnc_AnnTotPre-mm_Medina_2021through2080.xlsx" ) )
```

```python
AllAnnDF = pd.read_excel( InFiler, sheet_name="Simulated", header=0, index_col=0 )
```

```python
CN2All = AllAnnDF.loc[2051:].copy()
NRows = len( CN2All )
TotCols = len( list( CN2All.columns ) )
NRows, TotCols
```

```python
npCN2All = CN2All.to_numpy(dtype=np.float32).flatten()
npCN2All.shape
```

```python
fitP3 = sstats.pearson3.fit( npCN2All )
fitP3
```

```python
fP3Skew = fitP3[0]
fP3Loc = fitP3[1]
fP3Scale = fitP3[2]
```

```python
npPltCN2A = np.sort( npCN2All )
npPDFCN2A = sstats.pearson3.pdf( npPltCN2A, fP3Skew, loc=fP3Loc, scale=fP3Scale )
```

```python
rs1000 = sstats.pearson3.rvs( fP3Skew, loc=fP3Loc, scale=fP3Scale, size=1000 )
rs20 = sstats.pearson3.rvs( fP3Skew, loc=fP3Loc, scale=fP3Scale, size=20 )
```

```python
cBins = [ x * 50.0 for x in range(1, 52, 1) ]
```

```python
OutSVG = os.path.normpath( os.path.join( OUT_DIR1, "AnnPrecipPDF_w1000sample.svg" ) )
OutPNG = os.path.normpath( os.path.join( OUT_DIR1, "AnnPrecipPDF_w1000sample.png" ) )
OutPDF = os.path.normpath( os.path.join( OUT_DIR1, "AnnPrecipPDF_w1000sample.pdf" ) )
```

```python
Fig1 = plt.figure()
Fig1.set_size_inches(7.5, 4.8)
ax11 = Fig1.add_subplot(111)
s1000, bins1000, patches1000 = ax11.hist( rs1000, bins=cBins, color='xkcd:tangerine',
                                          label="1000 samples for 2060", density=True, alpha=0.75)
cn2All, = ax11.plot( npPltCN2A, npPDFCN2A, linestyle='-', linewidth=2.0, color='xkcd:ocean blue',
                     label="2051-2080 PDF", )
#ax11.set_xticks( IntMonths )
#ax11.set_yticks( YTicks )
#ax11.set_xticklabels( monTLabels, minor=False, rotation=0 )
ax11.set_xlabel('Annual Precipitation Depth ($mm$)', fontsize=10)
ax11.set_ylabel('Probability Density', fontsize=10)
#ax11.set_ylim( (YTicks[0], 450.0) )
#ax11.set_ylim( (YTicks[0], YTicks[len(YTicks)-1]) )
ax11.tick_params(axis='both', which='major', labelsize=10)
ax11.yaxis.set_major_formatter( mpl.ticker.StrMethodFormatter( "{x:,.4f}" ) )
ax11.xaxis.set_major_formatter( mpl.ticker.StrMethodFormatter( "{x:,.0f}" ) )
ax11.legend( loc='upper right', fontsize=9, ncol=1 )
ax11.grid(visible=True, which='major', axis='y' )
Fig1.savefig( OutSVG, dpi=600.0 )
Fig1.savefig( OutPNG, dpi=600.0 )
Fig1.savefig( OutPDF, dpi=600.0 )
```
